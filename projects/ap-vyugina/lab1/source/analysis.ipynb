```python
import pandas as pd
pd.set_option('display.max_colwidth', None)
```

```python
stats = pd.read_csv("stats.csv", 
                    names=["task", "model", "params", "total_duration, ns", "load_duration, ns", 
                           "prompt_eval_count", "prompt_eval_duration, ns", "eval_count", "eval_duration, ns"])
stats.head(5)
```

```python
data = pd.read_csv("res1.csv", names=["model", "params", "result"])
data
```

```python
stats_aggregated = stats.groupby(['task', 'model']).mean(numeric_only=True)
stats_aggregated["speed, token/s"] = stats_aggregated["eval_count"]/stats_aggregated["eval_duration, ns"] * 10**9
stats_aggregated["load_duration, s"] = stats_aggregated["load_duration, ns"] / 10**9
stats_aggregated["eval_duration, s"] = stats_aggregated["eval_duration, ns"] / 10**9

stats_aggregated["eval_count"] = stats_aggregated["eval_count"].astype(int)
stats_aggregated["prompt_eval_count"] = stats_aggregated["prompt_eval_count"].astype(int)
# stats_aggregated
stats_aggregated[['load_duration, s', 'prompt_eval_count', 'eval_count', 'eval_duration, s', 'speed, token/s']]
```

```python
stats_by_model_aggregated = stats.groupby(['model']).mean(numeric_only=True)
stats_by_model_aggregated["speed, token/s"] = stats_by_model_aggregated["eval_count"]/stats_by_model_aggregated["eval_duration, ns"] * 10**9
stats_by_model_aggregated["load_duration, s"] = stats_by_model_aggregated["load_duration, ns"] / 10**9
stats_by_model_aggregated[["load_duration, s", "speed, token/s"]]
```
